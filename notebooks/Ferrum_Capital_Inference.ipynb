```python
import re
import joblib
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
model = joblib.load("linear_svm_model.pkl")
tfidf = joblib.load("tfidf_vectorizer.pkl")
```

```python
label_map = {
    0: "anger",
    1: "joy",
    2: "love",
    3: "sadness",
    4: "fear",
    5: "surprise"
}
```

```python
def predict_emotion(text):
    cleaned = clean_text(text)
    vector = tfidf.transform([cleaned])
    pred_label = model.predict(vector)[0]
    return label_map[pred_label]
```

```python
sentence = "I didn't feel humiliated at all"
print("Predicted emotion:", predict_emotion(sentence))
```

```python
predict_emotion("I am extremely happy today")
```

```python
predict_emotion("I feel scared")
```

```python
predict_emotion("I miss her so much")
```

```python
predict_emotion("This makes me so angry")
```
